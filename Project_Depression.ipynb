```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['madrs1'], c='r', marker='o', label='Madrs1')
plt.xlabel('Age')
plt.ylabel('Madrs1')
plt.title("Age vs. Madrs1 Chart")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['madrs2'], c='b', marker='o', label='Madrs2')
plt.xlabel('Age')
plt.ylabel('Madrs2')
plt.title('Age vs. Madrs1 Chart')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
import numpy as np

# Eğitim ve test veri kümelerini ayırın
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Modeli seçin ve eğitin (örnek olarak Linear Regression kullanıyoruz)
model = LinearRegression()
model.fit(train_data[['days', 'age', 'madrs1']], train_data['madrs2'])

# Tahminleri yapın
tahminler = model.predict(test_data[['days', 'age', 'madrs1']])

# Tahminleri eksik verilere ekleyin (örneğin, madrs2 sütunu için)
test_data['madrs2'] = np.where(test_data['madrs2'].isna(), tahminler, test_data['madrs2'])

# Tahminleri eklenmiş veri çerçevesini kullanabilirsiniz
print(test_data)
```
